```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def oscillator(u0, mu=.95, nsteps=200):
    
    # Forward Euler solution to du/dt = mu + sin(u)
    
    dt = .1
    u = np.zeros(nsteps)
    t = np.zeros(nsteps)
    
    u[0] = u0 % (2*np.pi)    # make sure it's on [0, 2 pi)
    for j in range(0,nsteps-1):
        u[j+1] = ( u[j] + dt*(mu + np.sin(u[j])) ) % (2*np.pi)
        t[j+1] = t[j] + dt
    
    return u, t
        
```

```python
theta0 = [0, 1, 2, 3, 4, 5, 6]
N = 200
theta = np.zeros((N, len(theta0)))
fig, ax = plt.subplots(1, 2, figsize=(16,8))  
ax[0].grid(True)
ax[1].grid(True)

for n in range(0,len(theta0)):
    theta[:,n], t,  = oscillator(theta0[n], mu=.95, nsteps = N )
    ax[0].plot(t,theta[:,n])
    ax[1].plot(t,np.cos(theta[:,n]))

ax[0].set_xlabel('t')
ax[0].set_ylabel('theta')
ax[0].set_title('Trajectories d(theta)/dt = mu + sin(theta)')

ax[1].set_xlabel('t')
ax[1].set_title('cos(theta(t))')
```
