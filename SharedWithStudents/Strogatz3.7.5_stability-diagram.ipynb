```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the parametric equations we get in (d)
def r(x):
    return 2*x*(1/(1+x**2)**2)
def s(x):
    return r(x)*x - x**2/(1+x**2)
# create a discrete grid and calculate the values on it for visualization
xgrid = np.linspace(0,1,100)
rgrid = [0]*100
sgrid = [0]*100
for i in range(0,100):
    rgrid[i] = r(xgrid[i])
    sgrid[i] = s(xgrid[i])
```

```python
# Plot the stability diagram
plt.plot(rgrid,sgrid,label = '2 fixed points')
plt.xlabel('r')
plt.ylabel('s')
plt.text(0.34, 0.02, '3 fixed points', fontsize=12)
plt.text(0.2,0.06,'1 fixed point' , fontsize=12)
plt.legend()
plt.ylim(0,0.13)
plt.xlim(0,0.7)
plt.title('stability diagram')
plt.show()
```
