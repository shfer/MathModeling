Here we'll look at numerical solutions for the Van der Pol oscillator

$$
\ddot{x} + \mu(x^2-1)x + x = 0, \quad x(0) = x_0, \quad \dot{x}(0) = v_0
$$

Defining $v = \dot{x}$, this can be represented as the 2D nonlinear system

$$
\dot{\mathbf{y}} = \begin{bmatrix} v \\ \mu(1-x^2)v - x \end{bmatrix}, \quad \text{where}\quad
\mathbf{y} = \begin{bmatrix} x(t) \\ v(t) \end{bmatrix} 
$$


```python
from pylab import *
from scipy.integrate import ode
```

```python
def vdp(t, y, mu):
    """
    y is a 2 element array with the current state:  y[1] = x(t), y[2] = v(t)

    """
    x, v = y
    return [v, -x + mu*(1-x*x)*v]
```

```python
def runvdp(x0, v0, mu, T):

    Nt = T*100
    
    time = linspace(0, T, Nt)
    y = empty((Nt, 2))
    y[0] = [x0, v0]

    solver = ode(vdp)
    solver.set_integrator('dopri5')
    solver.set_f_params(mu)
    solver.set_initial_value(y[0], 0.)

    k = 1
    while solver.successful() and solver.t < T:
        solver.integrate(time[k])
        y[k] = solver.y
        k += 1
        
    fig, ax = subplots(1, 2, figsize=(16,4), gridspec_kw={'width_ratios': [1, 3]})   

    ax[0].plot(y[:,0],y[:,1], '-b')     
    ax[0].grid(True)
    ax[0].plot(y[0,0],y[0,1], 'oy')     # make a blue dot at the start
    ax[0].plot(y[-1,0],y[-1,1], 'ok')
    ax[0].set(xlabel='x', ylabel='v', title='Phase plot: x(t) vs. v(t)')

    ax[1].plot(time, y[:,0])
    ax[1].set(xlabel='t', ylabel='x', title='Position x(t)')
    ax[1].grid(True)

    return y, time, fig, ax
    
```

```python
# Large mu

mu = 10
T = 100

x0 = 3.
v0 = 20.

y, t, fig, ax = runvdp(x0, v0, mu, T)
```

```python
# Small mu

mu = .1
T = 100

x0 = 3.
v0 = 0.

y, t, fig, ax = runvdp(x0, v0, mu, T)

A = 2/sqrt(1+3*exp(-mu*t))
ax[1].plot(t, A*cos(t),'r-.')
ax[1].legend(["Numerical", "Two-timing solution $O(\epsilon^0)$"], loc="lower right")
```
