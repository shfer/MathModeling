```python
def cobweb(f, x0, Nsteps=100, x=linspace(0,1,1001), axlim=[-3, 3, -3, 3], ptime=.5, args={}):
    """
    Draws the cobweb and time series of the function f, with parameters args, starting
    from x0. Nsteps iterations are drawn. Change the interval by supplying the vector x of
    the points where f should be evaluated to plot its graph.

    """
    fig, ax = subplots(figsize=(8,8))  
    ax.plot(x, f(x, **args), '-b')   # plot the mapping function
    ax.plot(x, x, '--k')             # plot the line y=x
    ax.grid(True)
    xr = empty([Nsteps, 1])
    xr[0] = x0                       # save output at each n
    x = x0                           # overwrite array x with IC
    y = f(x, **args)
    ax.plot(x, y, 'by')              # make a blue dot at the start
    ax.plot([x, x], [0, y], '-r')    # plot the first vertical thread
    ax.axis(axlim)
    ax.draw
    for i in range(Nsteps):
        #print(x,y)
        ax.plot([x, y], [y, y], '-r') # horizontal thread
        x = y
        xr[i] = x
        y = f(x, **args)
        ax.plot([x, x], [x, y], '-r') # vertical thread
        display.display(fig)
        display.clear_output(wait=True)
        time.sleep(ptime)

    ax.plot(x, y, 'ok')
    
    # plot time series
    fig2, ax2 = subplots(figsize=(16,5))
    ax2.plot(xr,'.-k')
    ax2.axis([0, Nsteps, axlim[2], axlim[3]])
    ax2.grid(True)

    return xr
```

```python
def logistic(x, c=2, K=1):
    """
    logisitc map:  x_{n+1} = f(x_n) = (c+1-c*x_n/K)*x_n
    """
    return (c+1-c*x/K)*x

def cubic(x, c=3):
    """
    cubic map:  x_{n+1} = f(x_n) = c*x_n-x_n^3
    """
    return c*x-x*x*x
```
